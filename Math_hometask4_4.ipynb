##### Решите аналитически и потом численно (в программе) уравнение, зависящее от параметра а: 
##### sin(а*x)=0
##### при условии: 0.01<a<0.02, 100<х<500.
##### Т.е. надо найти решение х как функцию параметра а - построить график x=x(а).
##### Если численным методом не получается найти все ветви решения x(а), то отыщите хотя бы одну.
 __________________________________________________________________________________

#### Если $(a > 0) & (x > 0)$, то $a*x > 0$, то есть 
#### $a*x = pi + k*pi = pi*(1 + k) = pi*k$, где k = ±1, ±2, ±3, ..., ±int(inf)
#### для последовательного решения задачи пока можно ограничиться $a*x = pi$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(100, 500, 400)

plt.figure(figsize=(6, 6))
plt.plot(np.pi / x, x)
plt.xlim(0.01, 0.02)
plt.xlabel('a')
plt.ylabel('x')
plt.grid()
```

##### Из графика видно, что пределы 100<х<500 избыточны, чего нельзя сказать о пределах 0.01<a<0.02.
##### Ответ задачи можно сформулировать так: 
##### $a * x * k$, где $(0.01<a<0.02)$, $x = \frac{pi}{a}, k = ±1, ±2, ±3, ..., ±int(inf)$
