Постройте на одном графике две кривые y(x) для функции переменной y(k, x)=cos(k*x), взяв для одной кривой значение k=1, а для другой - любое другое k, не равное 1.

```python
# подключение внешних библиотек
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-20, 20, 200)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(x, np.cos(x*1), '-b')
plt.plot(x, np.cos(x*1.2), '--g')
plt.xlabel('y(x)')
plt.ylabel('cos(x)')
plt.show()
```
