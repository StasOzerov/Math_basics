### Задание 3
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
def draw_round(R: float, x0: float, y0: float, dots: int):
    """
    Функция построения окружности.
    R - радиус
    x0 - смещение по оси абсцис
    y0 - смещение по оси ординат
    dots - кол-во точек отрисовки окружности / 2
    """
    xlim_plus = x0 + 1.5 * R
    xlim_minus = x0 - 1.5 * R
    ylim_plus = y0 + 1.5 * R
    ylim_minus = y0 - 1.5 * R 
    
    x = np.linspace(-R, R, dots)
    y = np.sqrt(R ** 2 - x ** 2)
      
    plt.figure(figsize=(8, 8))
    plt.plot([xlim_minus, xlim_plus], [0, 0])
    plt.plot([0, 0], [ylim_minus, ylim_plus])
    plt.scatter(x0, y0)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(xlim_minus, xlim_plus)
    plt.ylim(ylim_minus, ylim_plus)
    plt.grid()
    
    plt.plot(x + x0, y0 + y)
    plt.plot(x + x0, y0 - y)


# Радиус окружности
R = 15

# Смещение по оси абсцис
x0 = 5

# Смещение по оси ординат
y0 = 10

# Кол-во точек отрисовки окружности / 2
dots = 50

draw_round(R, x0, y0, dots)
```

```python
def draw_elips(a: float, b: float, x0: float, y0: float, dots: int):
    """
    Функция построения элипса.
    a - половина длины элипса по оси абсцис
    b - половина длины элипса по оси ординат
    x0 - смещение по оси абсцис
    y0 - смещение по оси ординат
    dots - кол-во точек отрисовки элипса / 2
    """
    r_max = a if a > b else b
    xlim_plus = x0 + 1.5 * r_max
    xlim_minus = x0 - 1.5 * r_max
    ylim_plus = y0 + 1.5 * r_max
    ylim_minus = y0 - 1.5 * r_max 
    
    x = np.linspace(-a, a, dots)
    y = b * np.sqrt(1 - (x ** 2) / (a ** 2))
    
    plt.figure(figsize=(8, 8))
    plt.plot([xlim_minus, xlim_plus], [0, 0])
    plt.plot([0, 0], [ylim_minus, ylim_plus])
    plt.scatter(x0, y0)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(xlim_minus, xlim_plus)
    plt.ylim(ylim_minus, ylim_plus)
    plt.grid()
    
    plt.plot(x + x0, y0 + y)
    plt.plot(x + x0, y0 - y)


# Половина длины элипса по оси абсцис
a = 5

# Половина длины элипса по оси ординат
b = 3

# Смещение по оси абсцис
x0 = 2

# Смещение по оси ординат
y0 = -4

# Кол-во точек отрисовки элипса / 2
dots = 50

draw_elips(a, b, x0, y0, dots)
```

```python
def draw_hyperbole(a: float, b: float, x0: float, y0: float, dots: int):
    """
    Функция построения гиперболы.
    a - половина расстояния между двумя кривыми гиперболы
    b - крутизна гиперболы
    x0 - смещение по оси абсцис
    y0 - смещение по оси ординат
    dots - кол-во точек отрисовки гиперболы / 4
    """
    r_max = a if a > b else b
    dots = int(abs(dots))
    xlim_plus = x0 + 5 * r_max
    xlim_minus = x0 - 5 * r_max
    ylim_plus = y0 + 5 * r_max
    ylim_minus = y0 - 5 * r_max 
    
    x_pos = np.linspace(a, a * 5, dots)
    x_neg = np.linspace(-(a * 5), -a, dots)
    y_pos = b * np.sqrt((x_pos ** 2) / (a ** 2) - 1)
    y_neg = b * np.sqrt((x_neg ** 2) / (a ** 2) - 1)
    
    plt.figure(figsize=(8, 8))
    plt.plot([xlim_minus, xlim_plus], [0, 0])
    plt.plot([0, 0], [ylim_minus, ylim_plus])
    plt.scatter(x0, y0)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(xlim_minus, xlim_plus)
    plt.ylim(ylim_minus, ylim_plus)
    plt.grid()
    
    plt.plot(x_pos + x0, y0 + y_pos)
    plt.plot(x_pos + x0, y0 - y_pos)
    plt.plot(x_neg + x0, y0 + y_neg)
    plt.plot(x_neg + x0, y0 - y_neg)   
    

# Половина расстояния между двумя кривыми гиперболы
a = 1

# Крутизна гиперболы
b = 2

# Смещение по оси абсцис
x0 = 3

# Смещение по оси ординат
y0 = 3

# Кол-во точек отрисовки гиперболы / 4
dots = 50

draw_hyperbole(a, b, x0, y0, dots)
```
