1) Напишите код, который будет переводить полярные координаты в декартовы.

```python
def polar_to_cartesian(R: float, alpha: float) -> float:
    """
    Функция перевода из полярной системы координат в декартову.
    R - значение радиуса
    alpha - значение угла (от 0 до 2*Пи c периодом 2*Пи)
    """
    
    x = R * np.cos(alpha)
    y = R * np.sin(alpha)

    return x, y


R = 7
alpha = 10
x, y = polar_to_cartesian(R, alpha)
print(f'[{x}, {y}]')
```

2) Напишите код, который будет рисовать график окружности в полярных координатах.

```python
def draw_round_in_polar(R: float, x0: float, y0: float, dots: int):
    """
    Функция построения окружности.
    R - радиус
    x0 - смещение по оси абсцис
    y0 - смещение по оси ординат
    dots - кол-во точек отрисовки окружности / 2
    """
    xlim_plus = x0 + 1.5 * R
    xlim_minus = x0 - 1.5 * R
    ylim_plus = y0 + 1.5 * R
    ylim_minus = y0 - 1.5 * R 
    
    alpha = np.linspace(0, np.pi * 2, dots)
    x = R * np.cos(alpha)
    y = R * np.sin(alpha)    
    
    plt.figure(figsize=(8, 8))
    plt.plot([xlim_minus, xlim_plus], [0, 0])
    plt.plot([0, 0], [ylim_minus, ylim_plus])
    plt.scatter(x0, y0)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(xlim_minus, xlim_plus)
    plt.ylim(ylim_minus, ylim_plus)
    plt.grid()
    
    plt.plot(x + x0, y0 + y)
    plt.plot(x + x0, y0 - y)


# Радиус окружности
R = 10

# Смещение по оси абсцис
x0 = 3

# Смещение по оси ординат
y0 = 2

# Кол-во точек отрисовки окружности / 2
dots = 50

draw_round_in_polar(R, x0, y0, dots)
```

3) Напишите код, который будет рисовать граик отрезка прямой линии в полярных координатах.

```python
def polar_to_cartesian(R: float, alpha: float) -> float:
    """
    Функция перевода из полярной системы координат в декартову.
    R - значение радиуса
    alpha - значение угла (от 0 до 2*Пи c периодом 2*Пи)
    """
    
    x = R * np.cos(alpha)
    y = R * np.sin(alpha)

    return x, y


x1, y1 = polar_to_cartesian(3, 2)
x2, y2 = polar_to_cartesian(1, 1)

lim = max(list(map(abs, [x1, x2, y1, y2]))) + 1

plt.figure(figsize=(8, 8))
plt.plot([x1, x2], [y1, y2])
plt.scatter(x1, y1)
plt.scatter(x2, y2)

plt.plot([-lim, lim], [0, 0])
plt.plot([0, 0], [-lim, lim])
plt.xlabel('x') 
plt.ylabel('y')
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)
plt.grid()
    
```

##### 1) Решите систему уравнений:
$
\begin{cases}
y = x^2 – 1\\
exp(x) + x∙(1 – y) = 1
\end{cases}
$

```python
x = np.linspace(-20, 20, 100)

plt.figure(figsize=(8, 8))
plt.plot(x, x ** 2 - 1)
plt.plot(x, 1 - (1 - np.exp(x)) / x)
plt.xlim(-10, 10)
plt.ylim(-20, 50)
plt.grid()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x ** 2 + 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (-2.5, 5))
print(f'x1 = {x1}, y1 = {y1}')

x2, y2 = fsolve(equations, (2.5, 5))
print(f'x2 = {x2}, y2 = {y2}')

x3, y3 = fsolve(equations, (4.5, 20))
print(f'x3 = {x3}, y3 = {y3}')
```
