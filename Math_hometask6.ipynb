### 1 Задание (теорема сложения)
Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.


Формула сложения вероятностей для независимых событий P(A + B) = P(A) + P(B)

A - выпадет 2, B - выпадет 5

P(2) = 1/6, P(5) = 1/6

P(2) + P(5) = 1/3

### 2 Задание (теорема умножения)
Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

Формула вероятности для зависимых событий P(AB) = P(A)*P(B/A)

A - выпадет 2, B/A - после 2 выпадет 5

```python
p = 1/6 * 1/5
```

```python
print('P(AB) =', p)
```

### 7 Задание** (необязательное) 
Чёрный куб покрасили снаружи белой краской, затем разрезали на 27 одинаковых маленьких кубиков и как попало сложили из них большой куб. С какой вероятностью все грани этого куба будут белыми?


Решение:
    
У куба все стороны равны, длина стороны допустим 'a', тогда объем куба будет a*a*a = a**3 . 

Если мы распилим куб на 27 ровных кусочков, получится что объем каждого будет равен a**3 / 27, 

или a**3 / 3**3, т.е. каждая сторона маленького кубика будет равна a/3, и если из всех маленьких кубиков сложить один большой,

окажется что каждая квадратная сторона большого куба будет разделена на 9 маленьких одинаковых квадратов, 

и так на каждой из 6-ти сторон куба. 

У 8-ми кубиков, образующих пики большого куба 3 из 6 сторон окрашены в белый, 

а значит вероятность правильной позиции каждого такого кубика относительно себя равна 3/6, 

результирующая вероятность что все 8 кубиков в правильной позиции относительно себя равна (3/6)**8.

У 12-ти кубиков, находящихся посередине каждой грани, образуемой соединением двух строн большого куба, 2 из 6 сторон окрашены в белый,

а значит вероятность правильной позиции каждого такого кубика относительно себя равна 2/6,

результирующая вероятность что все 12 кубиков в правильной позиции относительно себя равна (2/6)**12.

У 6-ти кубиков в центре каждой стороны большого куба только 1 из 6 сторон окрашена в белый,

а значит вероятность правильной позиции каждого такого кубика относительно себя равна 1/6,

результирующая вероятность что все 6 кубиков в правильной позиции относительно себя равна (1/6)**6.

Один единственный кубик находящийся в центре большого куба остается неокрашенным,

а значит вероятность правильной позиции такого кубика относительно себя равна (1/1)**1.

А теперь, согласно формуле умножения вероятностей и формуле комбинаторики о перестановке

получаем вот такой громоздкий ответ:

$$P_{общ} = \frac{1}{27!} * (\frac{3}{6})^8 * (\frac{2}{6})^{12} * (\frac{1}{6})^6$$
